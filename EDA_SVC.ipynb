## 결측치 확인

```python
def check_missing(df):
    """주어진 데이터프레임의 결측치 확인 함수
    
    Args:
        df (pd.DataFrame): 검사할 데이터프레임.
    """
    for col in df.columns:  # 모든 열을 반복하여 검사
        missing_count = df[col].isnull().sum()  # 각 열의 결측치 개수를 계산
        if missing_count > 0:  # 결측치가 존재하는 경우
            # 결측치 개수와 데이터 타입을 출력
            print(f'{col} column missing values: {missing_count}, datatype: {df[col].dtype}')  

# 함수 실행
check_missing(training_set)  # training_set 데이터프레임의 결측치를 확인
```

```python
# 그래프의 행과 열 수 설정
ncols = 2
nrows = 12

# 서브플롯 생성 및 크기 설정
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(20, 40))

# 히스토그램을 그릴 열 이름 리스트
col_names = [
    'DroppedCalls', 'BlockedCalls', 'UnansweredCalls', 'CustomerCareCalls',
    'ThreewayCalls', 'ReceivedCalls', 'OutboundCalls', 'InboundCalls',
    'PeakCallsInOut', 'OffPeakCallsInOut', 'DroppedBlockedCalls', 
    'CallForwardingCalls', 'CallWaitingCalls', 'MonthsInService', 
    'UniqueSubs', 'ActiveSubs', 'Handsets', 'HandsetModels', 
    'CurrentEquipmentDays', 'RetentionCalls', 'RetentionOffersAccepted', 
    'ReferralsMadeBySubscriber', 'IncomeGroup', 
    'AdjustmentsToCreditRating'
]

def plot_hist(nrows, ncols, training_set, col_names):
    """주어진 데이터프레임의 여러 열에 대한 히스토그램을 그리는 함수
    
    Args:
        nrows (int): 서브플롯의 행 수
        ncols (int): 서브플롯의 열 수
        training_set (pd.DataFrame): 시각화할 데이터프레임
        col_names (list): 히스토그램을 그릴 열 이름 리스트
    """
    # 불필요한 축을 제거하기 위한 카운터
    counter = 0
    
    # 각 서브플롯에 대해 반복
    for i in range(nrows):
        for j in range(ncols):
            ax = axes[i][j]

            # 데이터가 있는 경우에만 플롯
            if counter < len(col_names):
                ax.hist(training_set[col_names[counter]], bins=20, label='{}'.format(col_names[counter]))
                ax.set_xlabel('Values')  # x축 레이블 설정
                ax.set_ylabel('Frequency')  # y축 레이블 설정
                leg = ax.legend(loc='upper right')  # 범례 설정
                leg.draw_frame(False)  # 범례 테두리 제거

            # 하단의 여분 축 제거
            else:
                ax.set_axis_off()

            counter += 1

    # 모든 히스토그램을 보여줌
    plt.show()

# 히스토그램 플롯 함수 호출
plot_hist(nrows, ncols, training_set, col_names)
```

```python
# 열 이름 리스트에서 'IncomeGroup' 열을 제거
col_names.remove('IncomeGroup')
```

```python
# CurrentEquipmentDays 열의 값이 0 이하인 경우 0으로 변환
# 제곱근 변환을 적용하기 위해서 이 열의 모든 값이 양수가 되도록 조정
training_set['CurrentEquipmentDays'] = training_set['CurrentEquipmentDays'].transform(lambda x: x if x > 0 else 0)
```

```python
def ohe_cat(ColName, training_set):
    encoder = OneHotEncoder(sparse_output=False)
    encoded_cat = encoder.fit_transform(np.array(training_set[ColName]).reshape(-1,1))
    encoded_cat = pd.DataFrame(encoded_cat, columns=[f'{ColName}_{cat}' for cat in sorted(training_set[ColName].unique())])
    training_set = training_set.drop(ColName, axis='columns')
    training_set = pd.concat([training_set, encoded_cat], axis=1)
    return training_set
transformed_set = ohe_cat('ServiceArea', transformed_set)
```

```python
# imputed_set을 imputed_transformed_set으로 복사하여 변환할 데이터프레임 생성
imputed_transformed_set = imputed_set

# last_col_names 리스트에 있는 각 열에 대해 아크쌍곡선 하이퍼볼릭 변환을 적용
for col in last_col_names:
    imputed_transformed_set[col] = imputed_transformed_set[col].transform(lambda x: np.arcsinh(x))

# 서브플롯의 열과 행 수 설정
ncols = 2
nrows = 3

# 서브플롯 생성: nrows와 ncols에 따라 크기를 지정
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(20, 40))

# 변환된 열들에 대한 히스토그램을 시각화하는 함수 호출
plot_hist(nrows, ncols, imputed_transformed_set, last_col_names)

# 결측치 확인 함수 호출: imputed_transformed_set 데이터프레임에서 결측치를 체크
check_missing(imputed_transformed_set)
```

### 아크쌍곡선 하이퍼볼릭 변환이 MonthlyRevenue, MonthlyMinutes, TotalRecurringCharge 및 OverageMinutes 열을 상대적으로 잘 정규화했음
이는 로그, 아크코쉬, 제곱근 변환보다 훨씬 효과적이며, 이전의 수치 변수들과 유사한 결과를 보임
히스토그램에서 많은 0 값의 영향이 특히 OverageMinutes에서 뚜렷하게 나타남
하지만 이 열들은 유지할 것임

RoamingCalls 및 DirectorAssistedCalls 열의 경우, 변환이 정규화에는 효과적이지 않았지만 분포를 약간 개선함
RoamingCalls와 DirectorAssistedCalls 열은 특정 값(즉, 0)의 빈도가 매우 높으므로 이들 또한 범주형 변수로 변경할 것임


```python
# 'RoamingCalls' 및 'DirectorAssistedCalls' 열에 대해 반복
for col in ['RoamingCalls', 'DirectorAssistedCalls']:
    # 아크쌍곡선 하이퍼볼릭 변환을 역변환하여 원래 값으로 복원
    imputed_transformed_set[col] = imputed_transformed_set[col].transform(lambda x: np.sinh(x))
    
    # 각 열의 값이 10,000보다 많이 발생하는 경우에는 해당 값을 문자열로 변환하고,
    # 그렇지 않은 경우 'Other'로 그룹화
    imputed_transformed_set[col] = imputed_transformed_set[col].transform(lambda x: str(x) if imputed_transformed_set[col].value_counts()[x] > 10000 else 'Other')
    
    # 고유한 값이 2개인 경우, 레이블 인코딩을 수행
    if len(imputed_transformed_set[col].unique()) == 2:
        imputed_transformed_set[col] = label_encoder.fit_transform(imputed_transformed_set[col])
    # 그렇지 않은 경우, 원-핫 인코딩을 수행
    else:
        imputed_transformed_set = ohe_cat(col, imputed_transformed_set)

# 변환된 데이터프레임 출력
imputed_transformed_set
```

# 이탈 위험 고객 분류 모델 훈련

```python
# 독립 변수(X) 설정: 'Churn' 열을 제거하여 고객 특성 데이터만 포함
X = new_feature_set.drop('Churn', axis='columns')

# 종속 변수(y) 설정: 'Churn' 열을 사용하여 고객 이탈 여부를 레이블로 설정
y = new_feature_set['Churn']

# X 데이터프레임의 열 이름을 문자열 형식으로 변환
X.columns = X.columns.astype(str)

# 데이터 정규화를 위한 StandardScaler 객체 생성
scaler = StandardScaler()

# X 데이터를 스케일링하여 평균 0, 표준편차 1로 변환
X = scaler.fit_transform(X)
```

```python
# SVC(서포트 벡터 분류기) 모델 생성: RBF(가우시안) 커널을 사용하고, 랜덤 상태를 0으로 설정하여 재현 가능성 확보
classifier = SVC(kernel='rbf', random_state=0)

# 훈련 데이터(X)와 레이블(y)을 사용하여 SVC 모델을 학습
classifier.fit(X, y)
```

```python
# SVC 모델의 교차 검증을 수행하여 정확도를 평가
# estimator: 사용할 모델, X: 독립 변수, y: 종속 변수, cv: 교차 검증을 위한 폴드 수(여기서는 5), n_jobs=-1: 모든 CPU 코어를 사용
accuracies = cross_val_score(estimator=classifier, X=X, y=y, cv=5, n_jobs=-1)

# 평균 정확도 출력: 교차 검증 결과의 평균 정확도를 백분율로 포맷하여 출력
print("Accuracy: {:.2f}%".format(accuracies.mean() * 100))

# 정확도 표준 편차 출력: 교차 검증 결과의 표준 편차를 백분율로 포맷하여 출력
print("Accuracy Standard Deviation: {:.2f}%".format(accuracies.std() * 100))
```

```python
import xgboost as xgb
from sklearn.model_selection import cross_val_score
from sklearn.metrics import precision_score, recall_score

# 데이터 분리
X = new_feature_set.drop('Churn', axis='columns')
y = new_feature_set['Churn']

# feature 이름을 모두 문자열로 변환
X.columns = X.columns.astype(str)

# 표준화
scaler = StandardScaler()
X = scaler.fit_transform(X)

# XGBoost 모델 생성
xgb_model = xgb.XGBClassifier(random_state=0)

# 교차 검증으로 모델 성능 확인
accuracies = cross_val_score(estimator=xgb_model, X=X, y=y, cv=5, n_jobs=-1)
print(f"Accuracy: {accuracies.mean()*100:.2f}%")
print(f"Accuracy Standard Deviation: {accuracies.std()*100:.2f}%")

# 모델 학습
xgb_model.fit(X, y)

# 예측
y_pred = xgb_model.predict(X)

# Precision, Recall 계산
precision = precision_score(y, y_pred)
recall = recall_score(y, y_pred)
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
```

```python
from sklearn.model_selection import GridSearchCV

# 하이퍼파라미터 그리드 정의
param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7],
    'subsample': [0.8, 1],
    'colsample_bytree': [0.8, 1]
}

# XGBoost 모델
xgb_model = xgb.XGBClassifier(random_state=0)

# GridSearchCV로 파라미터 탐색
grid_search = GridSearchCV(estimator=xgb_model, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)
grid_search.fit(X, y)

# 최적의 파라미터 출력
print(f"Best parameters: {grid_search.best_params_}")
best_xgb_model = grid_search.best_estimator_

# 최적 모델로 성능 평가
accuracies = cross_val_score(estimator=best_xgb_model, X=X, y=y, cv=5, n_jobs=-1)
print(f"Accuracy: {accuracies.mean()*100:.2f}%")
print(f"Accuracy Standard Deviation: {accuracies.std()*100:.2f}%")
```
